# ðŸ“¤ QualiVault: Export to Obsidian
**Goal:** Convert CSV transcripts into formatted Markdown files for your Obsidian Vault.

1. Reads `processing_recipe.yaml`.
2. Finds all completed transcripts.
3. Converts them to Markdown with inline metadata fields (for Dataview).
4. Saves them to `output/Obsidian_Vault/`.

```python
%load_ext autoreload
%autoreload 2
import yaml
from pathlib import Path
from qualivault.core import export_recipe
from qualivault import find_workspace_root

# 1. Load Config Paths
PROJECT_NAME = 'YOUR_PROJECT_NAME'  # <-- Change this to work on different projects
workspace_root = find_workspace_root(Path.cwd())
project_root = workspace_root / 'projects' / PROJECT_NAME
config_path = project_root / 'config.yml'

if not project_root.exists():
    raise FileNotFoundError(f"âŒ Project not found: {project_root}")
if not config_path.exists():
    raise FileNotFoundError(f"âŒ Config not found: {config_path}")

with open(config_path) as f:
    config = yaml.safe_load(f)

recipe_path = project_root / "processing_recipe.yaml"
output_dir = (project_root / config['paths']['output_base_folder']).resolve()

print(f"ðŸ“‚ Output Base: {output_dir}")
```

```python
# 2. Run Export
# Export is tolerant: it exports any transcripts it can find.
export_recipe(recipe_path, output_dir, config=config)
```
